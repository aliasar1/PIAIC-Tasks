```python
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score, classification_report
from sklearn.datasets import load_iris
import numpy as np

iris = load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

kmeans = KMeans(n_clusters=3)
kmeans.fit(X_train)

cluster_labels = kmeans.predict(X_test)

cluster_to_class = {0: np.argmax(np.bincount(y_train[kmeans.labels_ == 0])),
                    1: np.argmax(np.bincount(y_train[kmeans.labels_ == 1])),
                    2: np.argmax(np.bincount(y_train[kmeans.labels_ == 2]))}
predicted_labels = [cluster_to_class[label] for label in cluster_labels]

accuracy = accuracy_score(y_test, predicted_labels)*100
print("Accuracy:", accuracy)

report = classification_report(y_test, predicted_labels)
print("Classification Report:\n", report)
```
