```python
import numpy as np
```

```python
class SLP(object):
    def __init__(self, input_size, learning_rate = 0.5, epochs=285):
        self.weights = np.zeros(input_size+1)
        print("Creating weight vertoc, 1D with zeros:", self.weights)
        self.theta = 0.5
        self.learning_rate=learning_rate
        self.epochs = epochs
        
    def activation_function(self, input_value):
        return 1 if input_value >= 0 else 0
    
    def predict(self, input_value):
        z = self.weights.T.dot(input_value)
        a = self.activation_function(z)
        return a
    
    def perceptronLearning(self, user_input, desired_output):
        for j in range(self.epochs):
            for i in range(desired_output.shape[0]):
                x = np.insert(user_input[i], 0,1)
                y = self.predict(x)
                e = desired_output[i] - y
                print("Error:", e)
                print("Predicted Output:", y)
                self.weights = self.weights * self.learning_rate * x * e
                
user_input = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1],
])
desired_output = np.array([0,1,1,1])
print("User input:\n", user_input)
print("Desired Output:", desired_output)
slp = SLP(input_size=2)
slp.perceptronLearning(user_input, desired_output)
print("Input Weights with bias:", slp.weights)
print("Learning Rate:", slp.learning_rate)
print("Total Epochs:", slp.epochs)
```
