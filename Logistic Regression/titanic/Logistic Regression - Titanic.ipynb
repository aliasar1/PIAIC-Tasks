```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Age'] = df['Age'].fillna(29)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Sex'] = le.fit_transform(df['Sex']) 
```

```python
df.head()
```

```python
df = df.drop(['Name', 'Ticket'], axis='columns')
```

```python
df.head()
```

```python
df['Parch'].value_counts()
```

```python
df['Pclass'].value_counts()
```

```python
df['Cabin'].value_counts()
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=["Pclass","Embarked"])
df.head()
```

```python
df['Parch'].value_counts()
```

```python
df = df.drop(['SibSp', 'Parch', 'Cabin'], axis='columns')
```

```python
df.head()
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
col_names = ['PassengerId', 'Survived', 'Sex', 'Age', 'Fare', 'Pclass_1', 'Pclass_2', 'Pclass_3', 'Embarked_C',  'Embarked_Q',  'Embarked_S']
pima = df
```

```python
feature_cols = ['PassengerId', 'Sex', 'Age', 'Fare', 'Pclass_1', 'Pclass_2', 'Pclass_3', 'Embarked_C',  'Embarked_Q',  'Embarked_S']
X = pima[feature_cols]
y = pima.Survived
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
```

```python
y_pred
```

```python
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
class_names = [0,1]
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
sns.heatmap(pd.DataFrame(cnf_matrix), annot= True, cmap="YlGnBu", fmt='g')
ax.xaxis.set_label_position("top")
plt.title('Confusion Matrix', y=1.1)
plt.ylabel('Actual')
plt.xlabel('Predicted')
```

```python
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
print("Precision:", metrics.precision_score(y_test, y_pred))
print("Recall:", metrics.recall_score(y_test, y_pred))
```

```python
y_pred_proba = logreg.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr, tpr, label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```
