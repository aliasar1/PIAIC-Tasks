```python
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score, classification_report
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data_student.csv')
df.head(5)
```

```python
df.rename(columns={' UNS': 'UNS'}, inplace=True)
print(df)
```

```python
df['UNS'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['UNS'] = le.fit_transform(df['UNS'])
df
```

```python
target = df['UNS']
target
```

```python
inputs = df.drop(['UNS'], axis='columns')
inputs
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.3, random_state=1)

kmeans = KMeans(n_clusters=3)
kmeans.fit(X_train)

cluster_labels = kmeans.predict(X_test)

cluster_to_class = {0: np.argmax(np.bincount(y_train[kmeans.labels_ == 0])),
                    1: np.argmax(np.bincount(y_train[kmeans.labels_ == 1])),
                    2: np.argmax(np.bincount(y_train[kmeans.labels_ == 2]))}
predicted_labels = [cluster_to_class[label] for label in cluster_labels]

accuracy = accuracy_score(y_test, predicted_labels)*100
print("Accuracy:", accuracy)

report = classification_report(y_test, predicted_labels)
print("Classification Report:\n", report)
```
